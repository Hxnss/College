```python
import requests
from bs4 import BeautifulSoup

# Base URL for Data Science jobs in Indonesia (September 2024)
root = 'https://id.jobstreet.com/id/Data-Science-jobs?page='

# Lists to store extracted data
Job = []
Company = []
Location = []
Salary = []
JobAdd = []


for page in range(1, 5):
    url = requests.get(root + str(page))
    soup = BeautifulSoup(url.text, 'html.parser')


    articles = soup.find_all("article", {"class":"_8i9ct20 _8i9ct21 nbwpkd7i nbwpkd6e nbwpkd9q nbwpkd8m nbwpkdh nbwpkd66 nbwpkd5e _1cnegorb _1cnegor9 _1cnegora gs2fam18 gs2fam1b nbwpkd32 nbwpkd35"})

    for article in articles:

        job_title = article.find("div", {"class":"_8i9ct20 nbwpkd5g nbwpkd52"}).text.strip()
        Job.append(job_title)


        company_name = article.find("span", {"class":"_8i9ct20 nbwpkd4y gs2fam0 gs2fam1 gs2fam21 _1708b944 gs2fama"}).text.strip()
        Company.append(company_name)

        location_place = article.find("span", {"class":"_8i9ct20 nbwpkd4y gs2fam0 gs2fam1 gs2fam21 _1708b944 gs2fam7"}).text.strip()
        Location.append(location_place)

        salary_element = article.find("span", {"class":"_8i9ct20 _1ujqdkp2 nbwpkd4y nbwpkd0 nbwpkdr _1ujqdkp4"})
        if salary_element:
            salary = salary_element.text.strip()
        else:
            salary = "Not available"
        Salary.append(salary)

        job_added = article.find("span", {"class":"_8i9ct20 nbwpkd4y gs2fam0 gs2fam1 gs2fam22 _1708b944 gs2fam7"}).text.strip()
        JobAdd.append(job_added)

```

```python
len(Job), len(Company), len(Salary), len(Location), len(JobAdd)
```

```python
import pandas as pd
df = pd.DataFrame(
    {'Job':Job,
     'Perusahaan':Company,
     'Location':Location,
     'Salary':Salary,
     'Job_added':JobAdd,
     })
df.head()
```
